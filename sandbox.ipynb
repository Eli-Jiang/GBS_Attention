```python
%load_ext autoreload
%autoreload 2

import torch
# 1. еҜје…ҘдҪ еҲҡеҲҡеҶҷеҘҪзҡ„ Attention
# еҰӮжһңдҪ ж”ҫеңЁдәҶ src/model/transformer/attention.py йҮҢпјҡ
from src.model.gbs_attention import GBSAttention  # жҲ–иҖ…дҪ иҮӘе®ҡд№үзҡ„зұ»еҗҚ
from src.model.transformer import TimeSeriesTransformer  # еҰӮжһңдҪ жңүдёҖдёӘе®Ңж•ҙзҡ„ Transformer жЁЎеһӢ

# 2. жЁЎжӢҹдёҖдёӘиҫ“е…Ҙеј йҮҸ
# еҒҮи®ҫ Batch еӨ§е°Ҹдёә 2пјҢеәҸеҲ—й•ҝеәҰпјҲжҜ”еҰӮ 4 дёӘйҮҸеӯҗжҜ”зү№/иҠӮзӮ№/еҚ•иҜҚпјүпјҢзү№еҫҒз»ҙеәҰдёә 16
seq_len = 6
d_model = 6

x = torch.randn(seq_len, d_model)
print("иҫ“е…Ҙз»ҙеәҰ (X):", x.shape)

# 3. е®һдҫӢеҢ–дҪ зҡ„ Attention жЁЎеқ—
attention_layer = GBSAttention()  # ж №жҚ®дҪ зҡ„е®һзҺ°и°ғж•ҙеҸӮж•°

# 4. еүҚеҗ‘дј ж’ӯ
try:
    output = attention_layer.forward(x)
    print("рҹҺү иҝҗиЎҢжҲҗеҠҹпјҒ")
    print("иҫ“еҮәз»ҙеәҰ (Output):", output.shape)
    
    # ж ёеҝғж ЎйӘҢпјҡиҫ“еҮәз»ҙеәҰеҝ…йЎ»е’Ңиҫ“е…Ҙз»ҙеәҰе®Ңе…ЁдёҖиҮҙ [Batch, Seq_len, d_model]
    assert output.shape == x.shape, "з»ҙеәҰдёҚеҢ№й…ҚпјҒиҫ“е…Ҙе’Ңиҫ“еҮәзҡ„ Shape еә”иҜҘжҳҜдёҖж ·зҡ„гҖӮ"
    print("вң… з»ҙеәҰж ЎйӘҢйҖҡиҝҮпјҒ")
    print("иҫ“еҮәеј йҮҸ (Output):", output)
except Exception as e:
    print("вқҢ иҝҗиЎҢеҮәй”ҷдәҶпјҢжҠҘй”ҷдҝЎжҒҜеҰӮдёӢпјҡ")
    print(e)

if __name__ == "__main__":
    # жЁЎжӢҹи¶…еҸӮж•°
    batch_size = 1
    seq_len = 6      # иҜ»иҝҮеҺ» 96 дёӘж—¶й—ҙзӮ№
    pred_len = 2     # йў„жөӢжңӘжқҘ 24 дёӘж—¶й—ҙзӮ№
    c_in = 1          # еҚ•еҸҳйҮҸж—¶еәҸиҫ“е…Ҙ
    c_out = 1         # еҚ•еҸҳйҮҸж—¶еәҸиҫ“еҮә
    
    # 1. еҲқе§ӢеҢ–е®Ңж•ҙжЁЎеһӢ
    model = TimeSeriesTransformer(
        c_in=c_in, 
        c_out=c_out, 
        seq_len=seq_len, 
        pred_len=pred_len,
        d_model=6,   # жөӢиҜ•з”Ёе°ҸдёҖзӮ№зҡ„з»ҙеәҰпјҢи·‘еҫ—еҝ«
        n_heads=1, 
        n_layers=2
    )
    
    # 2. жһ„йҖ дёҖдёӘйҡҸжңәзҡ„иҫ“е…Ҙж•°жҚ® [Batch_Size, seq_len, c_in]
    dummy_input = torch.randn(batch_size, seq_len, c_in)
    
    # 3. еүҚеҗ‘дј ж’ӯ
    print("--- ејҖе§ӢйҖҡз”өжөӢиҜ• ---")
    try:
        dummy_output = model(dummy_input)
        print("рҹҺү жөӢиҜ•жҲҗеҠҹпјҒжЁЎеһӢжҲҗеҠҹи·‘йҖҡгҖӮ")
        print("жңҹжңӣиҫ“е…Ҙз»ҙеәҰ: [1, 6, 1]  --> е®һйҷ…иҫ“е…Ҙз»ҙеәҰ:", dummy_input.shape)
        print("жңҹжңӣиҫ“еҮәз»ҙеәҰ: [1, 2, 1]  --> е®һйҷ…иҫ“еҮәз»ҙеәҰ:", dummy_output.shape)
    except Exception as e:
        print("Error: ")
        print(e)
```
